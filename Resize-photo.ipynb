```python
import cv2
from PIL import Image
import os

def resize_image(img, width=720, height=480):
    # Resize the image
    resized_img = img.resize((width, height))
    return resized_img
if __name__ == "__main__":
    # Path to the folder containing the original photos
    original_folder_path = r"C:\Users\Salah\Downloads\GP\data\38"

    # Path to save the resized photos
    resized_folder_path = r"C:\Users\Salah\Downloads\GP\redata\38"

    # Create the resized folder if it does not exist
    if not os.path.exists(resized_folder_path):
        os.makedirs(resized_folder_path)

    # Loop through all the files in the original folder
    for filename in os.listdir(original_folder_path):
        if filename.endswith(".jpg") or filename.endswith(".png"): # Check if the file is an image file
            # Load the original photo
            original_photo_path = os.path.join(original_folder_path, filename)
            img = Image.open(original_photo_path)

            # Resize the photo
            resized_photo = resize_image(img)

            # Path to save the resized photo
            resized_photo_path = os.path.join(resized_folder_path, filename)

            # Save the resized photo
            resized_photo.save(resized_photo_path)
            print(f"Resized photo saved as '{filename}'")
```
